```python
#Bài 1:
#Import thư viện
from PIL import Image
import numpy as np

#Đọc ảnh
img = Image.open('bird.png')
img.show()  # Hiển thị ảnh gốc

#Chuyển ảnh thành mảng NumPy
img_array = np.array(img)

#Tạo các ảnh màu riêng biệt

#Kênh Red
red_array = np.zeros_like(img_array)
red_array[:, :, 0] = img_array[:, :, 0]
red_img = Image.fromarray(red_array)
red_img.save('bird_red.png')

#Kênh Green
green_array = np.zeros_like(img_array)
green_array[:, :, 1] = img_array[:, :, 1]
green_img = Image.fromarray(green_array)
green_img.save('bird_green.png')

#Kênh Blue
blue_array = np.zeros_like(img_array)
blue_array[:, :, 2] = img_array[:, :, 2]
blue_img = Image.fromarray(blue_array)
blue_img.save('bird_blue.png')

red_img.show()
green_img.show()
blue_img.show()
```

```python
#Bài 2:
from PIL import Image
import numpy as np

#Nạp ảnh
img = Image.open('flower.jpeg')  #Đổi tên file nếu cần
img_array = np.array(img)

#Kiểm tra nếu ảnh có kênh alpha thì loại bỏ
if img_array.shape[2] == 4:
    img_array = img_array[:, :, :3]

#Các tổ hợp hoán đổi kênh màu
# Gốc: [R, G, B]
# Hoán đổi: [G, B, R], [B, R, G], [R, B, G], [G, R, B], [B, G, R]

swap_1 = img_array[:, :, [1, 2, 0]]  # G, B, R
swap_2 = img_array[:, :, [2, 0, 1]]  # B, R, G
swap_3 = img_array[:, :, [0, 2, 1]]  # R, B, G
swap_4 = img_array[:, :, [1, 0, 2]]  # G, R, B
swap_5 = img_array[:, :, [2, 1, 0]]  # B, G, R

#Chuyển về ảnh và lưu
Image.fromarray(swap_1).save('flower_GBR.png')
Image.fromarray(swap_2).save('flower_BRG.png')
Image.fromarray(swap_3).save('flower_RBG.png')
Image.fromarray(swap_4).save('flower_GRB.png')
Image.fromarray(swap_5).save('flower_BGR.png')
```

```python
#Bài 3:
from PIL import Image
import numpy as np
import colorsys

#Nạp ảnh RGB và chuyển sang NumPy
img = Image.open('bird.png').convert('RGB')
rgb_array = np.array(img)

#Chuyển từ RGB sang HSV (giá trị từ 0-1)
rows, cols, _ = rgb_array.shape
hsv_array = np.zeros_like(rgb_array, dtype=np.float32)

for i in range(rows):
    for j in range(cols):
        r, g, b = rgb_array[i, j] / 255.0  # Chuẩn hóa
        h, s, v = colorsys.rgb_to_hsv(r, g, b)
        hsv_array[i, j] = [h, s, v]

#Tạo ảnh riêng cho H, S, V
#Hình ảnh đơn kênh được nhân lên 3 kênh để lưu được
h_img = np.uint8(hsv_array[:, :, 0] * 255)
s_img = np.uint8(hsv_array[:, :, 1] * 255)
v_img = np.uint8(hsv_array[:, :, 2] * 255)

#Chuyển thành ảnh RGB từ ảnh đơn kênh (xám)
h_rgb = Image.fromarray(np.stack([h_img]*3, axis=-1))
s_rgb = Image.fromarray(np.stack([s_img]*3, axis=-1))
v_rgb = Image.fromarray(np.stack([v_img]*3, axis=-1))

#Lưu ảnh
h_rgb.save('bird_H.png')
s_rgb.save('bird_S.png')
v_rgb.save('bird_V.png')

#(Tuỳ chọn) Hiển thị ảnh
h_rgb.show(title="Hue (H)")
s_rgb.show(title="Saturation (S)")
v_rgb.show(title="Value (V)")
```

```python
#Bài 4:
from PIL import Image
import numpy as np
import colorsys

# Bước 1: Nạp ảnh RGB
img = Image.open('flower.jpeg').convert('RGB')
rgb_array = np.array(img)

# Bước 2: Chuyển RGB sang HSV, áp dụng Hnew và Vnew
rows, cols, _ = rgb_array.shape
output_array = np.zeros_like(rgb_array)

for i in range(rows):
    for j in range(cols):
        r, g, b = rgb_array[i, j] / 255.0  # Chuẩn hóa
        h, s, v = colorsys.rgb_to_hsv(r, g, b)

        # Điều chỉnh H và V theo yêu cầu
        h_new = (1/3) * h
        v_new = (3/4) * v

        # Chuyển ngược về RGB
        r_new, g_new, b_new = colorsys.hsv_to_rgb(h_new, s, v_new)
        output_array[i, j] = [int(r_new * 255), int(g_new * 255), int(b_new * 255)]

# Bước 3: Tạo ảnh mới và lưu
new_img = Image.fromarray(output_array.astype(np.uint8))
new_img.save('flower_modified_HV.png')
new_img.show(title='Modified HSV Image')
```

```python
#Bài 5:
import cv2
import os

#Tạo thư mục đầu ra nếu chưa có
input_folder = 'Exercise'
output_folder = 'Exercise_filtered'

os.makedirs(output_folder, exist_ok=True)

#Duyệt qua từng ảnh trong thư mục
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        #Đọc ảnh
        img = cv2.imread(input_path)

        if img is None:
            print(f"Lỗi khi đọc ảnh: {filename}")
            continue

        #Áp dụng mean filter (bộ lọc trung bình 3x3)
        filtered_img = cv2.blur(img, (5, 5))  # Có thể thử (5, 5) nếu muốn mượt hơn

        #Lưu ảnh đã lọc
        cv2.imwrite(output_path, filtered_img)

print("Đã xử lý xong tất cả ảnh trong thư mục 'Exercise/'.")
```
